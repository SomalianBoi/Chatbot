```python
import pandas as pd
import os
```

```python
file_path = "D:\\Users\\User\\Desktop\\chatbot\\dataset_cleaned.csv"
data = pd.read_csv(file_path)
```

```python
data
```

```python
categories = data['Category'].unique()
```

```python
output_dir ="D:\\Users\\User\\Desktop\\chatbot"
os.makedirs(output_dir, exist_ok=True)
```

```python
for category in categories:
    df_category = data[data['Category'] == category]
    
    category_file_path = os.path.join(output_dir, f"{category}_dataset.csv")
    
    df_category.to_csv(category_file_path, index=False)
    print(f"Saved dataset for category '{category}' to: {category_file_path}")

print("All category datasets have been created.")
```
